```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('GLB.Ts_dSST.csv', header=1 )
df.head()
```

```python
df.drop(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec', 'D-N', 'DJF', 'MAM', 'JJA', 'SON'], axis = 1, inplace = True)
df.head()
```

```python
chart = alt.Chart(df).mark_point(filled = True, size = 60).encode(
    x = alt.X(field='Year', 
              title = "",
              scale = alt.Scale(
                  zero = False),
              axis = alt.Axis(tickCount = 10),
              type = 'quantitative'),
    y = alt.Y(field = 'J-D',
              title = "",
              axis = alt.Axis(tickCount = 10, format = '+'),
              type = 'quantitative'),
    color = alt.value("#FF8C00"))

line = pd.DataFrame({
    'Year': [1880, 2017],
    'Temp':  [-0.2, -0.2],
})

line = alt.Chart(line).mark_line(color= 'black').encode(
    x= 'Year',
    y= 'Temp',
)

text1 = alt.Chart({'values':[{'x': 2009, 'y': -0.1}]}).mark_text(lineBreak='\n',
    text=' HOTTER THAN THE\n1880-1899 AVERAGE'
).encode(
    x='x:Q', y='y:Q'
)

text2 = alt.Chart({'values':[{'x': 2013, 'y': -0.27}]}).mark_text(
    text='COLDER'
).encode(
    x='x:Q', y='y:Q'
)

text3 = alt.Chart({'values':[{'x': 1883, 'y': 0.9}]}).mark_text(lineBreak='\n',
    align='left',
    baseline='middle',
    dx=5,
    fontSize = 15,                                                            
    text='Annual Global Surface Temperature,\nRelative to Late 19th Century Average'
).encode(
    x='x:Q', y='y:Q'
)

points = df[df["Year"].isin([1904, 1944, 1998, 2014, 2015, 2017, 2016])]

points_chart = alt.Chart(points).mark_point().encode(
    x='x:Q',
    y='y:Q'
)

labels = points_chart.mark_text(
    align='left',
    baseline='middle',
    dx=7,
    size = 12
).encode(
    x = alt.X(title = "", field='Year', type = 'quantitative'),
    y = alt.Y(title = "", field = 'J-D', type = 'quantitative'),
    text='Year'
)

alt.layer(line, chart, text1, text2, text3, labels).properties(width = 950, height = 500, background = '#FFFFFF', padding = 25)

```
